**Task 1.1**

scalar valued functions 
$$

y = mx + b
$$

interval of  **x ∈ [−1, 1]**

```python
#import required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# assign slop,intersection and create the interval
m=1
b=2
x=np.linspace(-1,1,100)
```

```python
#calculation
y=m*x+b
```

```python
# Plot the solution
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='y = mx + b', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear function plot')
plt.legend()
plt.grid(True)
plt.show()
```
